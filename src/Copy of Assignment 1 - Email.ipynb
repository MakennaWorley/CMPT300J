# Assignment 1 - Email

Design a class of `Email` with following requirements:

- It has four attributes:
    - sender: the email address of the sender
    - recipients: the email addresss of the recipients
    - subject: the title of the email
    - body: the content of the email

- Special method `setitem` with key and value:
    - key is the attribuet and value is the content of the attribute
    - set the values of attributes using **match and case**
    - the value is a single string

- Special method `len`:
    - return the lenght of the string of the body. 

- Special method `eq`:
    - return True if the values of all four attributes of two emails are same. 
    - return False otherwise. 

- Special method `str`:
    - print out the email using **f-string** following this format:
    ```
    From: sender email address
    To: recipient email address, another recipient email address
    Subject: subject

    This is the email body.
    ```

````python
class Email:
    def __init__(self):
        self.sender = ''
        self.recipients = ''
        self.subject = ''
        self.body = ''

    def __setitem__(self, key, value):
        match key:
          case 'sender':
            self.sender = value
          case 'recipients':
            self.recipients = value
          case 'subject':
            self.subject = value
          case 'body':
            self.body = value

    def __len__(self):
        return len(self.body)

    def __eq__(self, other):
        return (self.sender, self.recipients, self.subject, self.body) == (other.sender, other.recipients, other.subject, other.body)

    def __str__(self):
        return f"From: {self.sender} \nTo: {self.recipients} \nSubject: {self.subject} \n \n{self.body}"
````

````python
# Test case 1
# Excepted output: 30

email = Email()

email['sender'] = 'jliang@westminstercollege.edu'
email['recipients'] = 'abc0101@westminstercollege.edu'
email['subject'] = 'Hello!'
email['body'] = 'This is your first assignment.'

print(len(email))
````

````python
# Test case 2
# Excepted output:
"""
From: jliang@westminstercollege.edu
To: abc0101@westminstercollege.edu
Subject: Hello!

This is your first assignment.
"""
print(email)
````

````python
# Test case 3
# Excepted output: True

email2 = Email()

email2['sender'] = 'jliang@westminstercollege.edu'
email2['recipients'] = 'abc0101@westminstercollege.edu'
email2['subject'] = 'Hello!'
email2['body'] = 'This is your first assignment.'

print(email == email2)
````
